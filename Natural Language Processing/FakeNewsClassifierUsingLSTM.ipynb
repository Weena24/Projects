```python
from keras.layers import Embedding #help to implement word2vec
from keras.preprocessing.sequence import pad_sequences #prepadding or postpadding
from keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from keras.layers import LSTM
from keras.layers import Dense
```

```python
### Vocabulary size - unique words present
voc_size=5000
```

### Onehot Representation
#### at which index are the words present

```python
messages=X.copy()
```

```python
messages['title'][1]
```

```python
messages
```

```python
#messages.reset_index(inplace=True)
```

```python
import nltk
import re #regular expression (regex)
from nltk.corpus import stopwords #(english stopwords)
```

```python
nltk.download('stopwords')
```

```python
### Dataset Preprocessing
from nltk.stem.porter import PorterStemmer ##stemming purpose
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i]) #apart from a-z and A-Z, remove special characters and replace with blank
    review = review.lower() #lower all title
    review = review.split() #get words by words

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')] #if word is not in stopwords english, stem it, if present in stopwords, ignore it
    review = ' '.join(review) #join all the review
    corpus.append(review) #create the corpus
```

```python
corpus
```

```python
corpus[1]
```

```python
onehot_repr=[one_hot(words,voc_size)for words in corpus]
onehot_repr
```

```python
corpus[1]
```

```python
onehot_repr[1]
```

### Embedding Representation
#### Word2vec representation

```python
#length of all sentences should be same
sent_length=20 #value should cover length of all the sentences
embedded_docs=pad_sequences(onehot_repr,padding='post',maxlen=sent_length) #add zeros at the end of the sentences #postpadding
print(embedded_docs)
```

```python
embedded_docs[1] #20 values - length of all sentences should be 20
```

```python
embedded_docs[0]
```

```python
## Creating model - apply embedding and LSTM layer
embedding_vector_features=40 ##features representation -each word at particular index will be converted by a vector of size 40, every word will be represented by a 40-dimensional vector
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features, input_length=sent_length)) #voc_size=5000,embedding_vector_features=40,input_length=20
model.add(LSTM(100)) #100 neurons
model.add(Dense(1,activation='sigmoid')) #output is binary, thus activation function is sigmoid
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy']) #binary output - binary_crossentropy
```

```python
model.build(input_shape=(None, sent_length))
print(model.summary())
```

```python
len(embedded_docs),y.shape
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

### Model Training

```python
### Finally Training
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

### Adding Dropout

```python
from tensorflow.keras.layers import Dropout
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3)) #30% of the LSTM while training will be disabled. also drop out reduces overfitting
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

### Performance Metrics And Accuracy

```python
y_pred=model.predict(X_test)
```

```python
y_pred=np.where(y_pred > 0.6, 1,0) ##AUC ROC Curve #1-fake news and 0-not fake news
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
