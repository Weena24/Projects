## MatplotLib Tutorial

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

```python
import matplotlib.pyplot as plt

%matplotlib inline # no need to type plt.show() everytime when doing the graphs 
```

```python
import numpy as np
```

```python
## Simple Examples
## arrange is used for creating an array, it has a start and a stop. 2D diagram - thus create 2 variables - x and y

x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
##plotting using matplotlib 

##plt scatter

plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png') #image will be saved in jupyter notebook folder or you can right-click and save image

```

```python
y=x*x # y=x*2 thus graph will be exponentially increasing 
```

```python
## plt plot

plt.plot(x,y,'r*',linestyle='dashed',linewidth=2, markersize=12) #r* displys the points as red and *
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram');
```

```python
## Creating Subplots

plt.subplot(2,2,1) #2 rows and 2 columns, 1st position ---> 1 diagram
plt.plot(x,y,'r--') #works with the 1st line - this diagram is positioned first (red dotted line diag)
plt.subplot(2,2,2) #2 rows and 2 columns, 2nd position ---> 2nd diagram (green dotted plot)
plt.plot(x,y,'g*--') #(green dotted plot)
plt.subplot(2,2,3) #3rd position 
plt.plot(x,y,'bo')
plt.subplot(2,2,4) #4th position
plt.plot(x,y,'go');
```

```python
x = np.arange(1,11) 
y = 3 * x + 5 
plt.title("Matplotlib demo") 
plt.xlabel("x axis caption") 
plt.ylabel("y axis caption") 
plt.plot(x,y) 
plt.show()
```

```python
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y) 
plt.show() 
```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1) #2 rows and 1 column - 1st index position 
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) #2nd index position
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

```python
## Bar plot

x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y)  # bar chart
plt.bar(x2, y2, color ='g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

## Histograms

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) #by default hist display 10 bins
#plt.hist(a, bins=20)
plt.title("histogram") # numbers falling between 10 to 20 ---> 1 #density is displayed on y axis
plt.show()
```

## Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)] # normal distribution, step size is 100 from 0 to 1...range is (1,4), 
#it will loop from 1st step - 0 to 1, 2nd step - 0 to 2, 3rd step - 0 to 3

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);  #vert=True means that the box plot is dislayed vertically, patch_artist=True displays colour of the box plot
```

## Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210] #these values are calculated as percentage and displayed in the pie chart
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0.2, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False) # explode means how far the slice gets moved from the pie chart, autopct percentage shows in float format

plt.axis('equal') #make sure the pie chart appears as a perfect cicle and is not distorted
plt.show()
```
