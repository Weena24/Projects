## Time Resampling 

```python
df_tesla.head()
```

```python
#rule A means year and frequency
df_tesla.resample(rule='A').max()['Open'].plot(); #maximum values based on year
```

```python
#quarterly start frequency 
df_tesla.resample(rule='QS').max()['High'].plot();
```

```python
#Business end frequency 
df_tesla.resample(rule='BA').max()['High'].plot();
```

```python
#Plot
df_tesla['Open'].resample(rule='BA').mean().plot(kind='bar');
```

```python
df_tesla['High'].rolling(10).mean() #rolling window of 10 days. rolling down by 10 steps then calculate mean, reason for NaN.
```

```python
df_tesla.tail()
```

```python
df_tesla['Open:30 days rolling']=df_tesla['Open'].rolling(31).mean()
```

```python
df_tesla.head()
```

```python
df_tesla[['Open','Open:30 days rolling']].plot(figsize=(12,5));
```
