### Feature Selection Techniques 

```python
import pandas as pd
```

```python
df=pd.read_csv('mobile_dataset.csv')
```

```python
df.head()
```

```python
df.duplicated()
```

### Univariate Selection

```python
X=df.iloc[:,:-1] #take all rows except last one ---> independent features
y=df['price_range'] # dependent variable
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.feature_selection import SelectKBest   #which features you want
from sklearn.feature_selection import chi2
```

```python
### Apply KBest algorithm
ordered_rank_features=SelectKBest(score_func=chi2,k=20)
ordered_feature=ordered_rank_features.fit(X,y)
```

```python
ordered_feature
```

```python
pd.DataFrame(ordered_feature.scores_,columns=['Score'])
```

```python
dfscores=pd.DataFrame(ordered_feature.scores_,columns=['Score'])
dfcolumns=pd.DataFrame(X.columns)
```

```python
features_rank=pd.concat([dfcolumns,dfscores],axis=1)
```

```python
features_rank.columns=['Features','Score']
features_rank
```

```python
#top 10 features wrt univariate selection
features_rank.nlargest(10,'Score')
```

### Feature Importance
This technique gives you a score for each feature of your data, the higher the score, more relevant it is

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
```

```python
model=ExtraTreesClassifier()
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

```python
ranked_feature=pd.Series(model.feature_importances_,index=X.columns)
```

```python
ranked_feature.nlargest(10).plot(kind='barh')
plt.show()
```

```python
### price is removed here
### 3G and 4G are highly correlated - both features are doing the same task - can drop one of them 
corr=df.iloc[:,:-1].corr()
top_features=corr.index
plt.figure(figsize=(20,20))
sns.heatmap(df[top_features].corr(),annot=True)
```

```python
#whichever feature has a correlation value of 0.8, more than 80%, drop that feature 
threshold=0.5
```

```python
### Remove the correlated features which are correlated among themselves (independent features)
#find and remove correlated features
def correlation(dataset, threshold):
    col_corr = set() #Set of all names of correlated columns --- create a set that is unique values 
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)): #iterating from y side and x side from heatmap
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: #if threshold is >0.5.... (0.5 is selected according to domain expert)
                colname = corr_matrix.columns[i] # getting the name of column
                col_corr.add(colname)
    return col_corr #correlated columns
```

```python
#highly correlated features displayed based on threshold value
correlation(df.iloc[:,:-1],threshold)
```
