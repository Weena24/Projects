## Black Friday Dataset EDA and Feature Engineering

### Cleaning and preparing the data for model training 

```python
## dataset link: https://www.kaggle.com/datasets/sdolezel/black-friday?select=train.csv

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Problem Statement

A retail company “ABC Private Limited” wants to understand the customer purchase behaviour (specifically, purchase amount) against various products of different categories. They have shared purchase summary of various customers for selected high volume products from last month. The data set also contains customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details (product_id and product category) and Total purchase_amount from last month.

Now, they want to build a model to predict the purchase amount of customer against various products which will help them to create personalized offer for customers against different products.

```python
#importing the dataset 
df_train=pd.read_csv('train.csv')
df_train.head()
```

```python
df_train.shape
```

```python
## import the test data ---> here output is not visible as it is the test data 
df_test=pd.read_csv('test.csv')
df_test.head()
```

```python
df_test.shape
```

```python
## Merge both train and test data
#pd.merge is when you add columns to the dataframe - vertical addition
#pd.concat is when you add rows to the dataframe - horizontal addition 
df=pd.concat([df_train, df_test], ignore_index=True)
df.head()
```

```python
# confirmation that train and test data were added
df.shape
```

```python
df.info()

#age is an object as it was inserted in range
```

```python
df.describe()
```

```python
#axis=1 is the column, inplace=True will drop the User_ID feature and update the df
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
#User_ID is gone 
df.head()
```

```python
#map function wil map wrt conditions given 
## Handling categorical feature Gender
df['Gender']=df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
## Handling categorical feature Age
##.unique is displaying how many categories there are in Age
df['Age'].unique()
```

```python
#Target Guiding ---> Model will better understand the maths
df['Age']=df['Age'].map({'0-17':1,'18-25':2, '26-35':3, '36-45':4, '46-50':5, '51-55':6, '55+':7})
```

```python
df.head()
```

```python
##fixing categorical City_Category - #pd.get_dummies(df['City_Category'],drop_first=True).astype(int)
##why drop_first=True ---> if i have 3 categories, 2 categories are enough to represent the 3 categories 
df_city=pd.get_dummies(df['City_Category'],drop_first=True).astype(int)
```

```python
df_city.head()
```

```python
df=pd.concat([df,df_city],axis=1)
df.head()
```

```python
##drop City_Category Feature
##inplace=True ---> update the df ---> make this permanent
df.drop(['City_Category'],axis=1,inplace=True)
```

```python
df.head()
```

```python
##Check missing values
##Purchase has missing values because part of it is the test data
df.isnull().sum()
```

```python
##Focus on replacing missing values 
##when we do .unique ---> we get a discrete feature
df['Product_Category_2'].unique()
```

```python
##value_counts gives the frequency the every number occurs 
df['Product_Category_2'].value_counts()
```

```python
#the display is a series
df['Product_Category_2'].mode()
```

```python
#put index to remove the series and display only mode value
df['Product_Category_2'].mode()[0]
```

```python
## Replace the missing values with mode 
## Cannot use mean as mean will create a new category 
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
##do same for Product_Category_3
df['Product_Category_3'].mode()[0]
```

```python
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
df['Product_Category_3'].isnull().sum()
```

```python
df.head()
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
df.head()
```

```python
##convert object into integers
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].astype(int)
```
