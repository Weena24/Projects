```python
df=pd.read_csv('mercedes.csv')
df.head()
```

```python
df=pd.read_csv('mercedes.csv', usecols=['X0','X1','X2','X3','X4','X5','X6'])
df.head()
```

```python
df['X0'].value_counts()
```

```python
df['X0'].unique()
```

```python
#First feature X0 is having 47 categories, X1 27 categories...etc
#Lots of categories ---> one hot encoding cannot be used 
for i in df.columns:
    print(len(df[i].unique()))
```

```python
##KDD Cup Orange Challenge ---> take top 10 categories for each feature and do one hot encoding, skip other features 
#Do same when having many features 
```

```python
#see top 10 most frequent variables for each feature. ascending is false as we are seeing the most frequennt categories 
df.X1.value_counts().sort_values(ascending=False).head(10)
```

```python
#see their indexes in the feature
df.X1.value_counts().sort_values(ascending=False).head(10).index
```

```python
lst_10=df.X1.value_counts().sort_values(ascending=False).head(10).index
```

```python
#convert the categories indexes into a list 
lst_10=list(lst_10)
```

```python
lst_10
```

```python
import numpy as np

for categories in lst_10:
    df[categories]=np.where(df['X1']==categories,1,0)
```

```python
lst_10.append('X1')
```

```python
#v is 1, t is 0 as it is not present in top 10 categories 
df[lst_10]
```
