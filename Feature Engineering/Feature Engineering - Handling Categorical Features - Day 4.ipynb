```python
#Renaming columns
train_set.columns=['Employment','Qualification','Status','Designation','Family title','Race','Sex','Country']
```

```python
train_set.head()
```

```python
for feature in train_set.columns[:]:
    print(feature,':',len(train_set[feature].unique()),'labels')
```

```python
train_set['Country'].value_counts()
```

```python
train_set['Country'].value_counts().to_dict()
```

```python
country_map=train_set['Country'].value_counts().to_dict()
```

```python
train_set['Country']=train_set['Country'].map(country_map)
train_set.head(20)
```

### Advantages
1. Easy to use
2. Not increasing any feature space

### Disadvantages
1. If the frequency of two features are same, it will provide the same weight. Not able to distinguish the proper information

### Target Guided Ordinal Encoding 

1.Ordering the labels according to the target

2. Replace the labels by the joint probability of being 1 or 0 in classification problem 

```python
import pandas
df=pd.read_csv('titanic.csv', usecols=['Cabin','Survived'])
df.head()
```

```python
df['Cabin'].fillna('Missing', inplace=True)
```

```python
df.head(20)
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
```

```python
df.head()
```

```python
df.Cabin.unique()
```

```python
#probability of person surviving from Cabin A is 0.46, from Cabin B - 0.74...etc
df.groupby(['Cabin'])['Survived'].mean()
```

```python
#sorting done wrt to probabilities
df.groupby(['Cabin'])['Survived'].mean().sort_values()
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels=df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels
```

```python
#D has the highest probability of survival, thus the rank is highest ~ 8
#k indicates the labels, i indicates the value starting from 0
#enumerate is like iterate 
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,0)}
ordinal_labels2
```

```python
df['Cabin_ordinal_labels']=df['Cabin'].map(ordinal_labels2)
df.head()
```

### Mean Encoding

```python
df.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
mean_ordinal=df.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
mean_ordinal
```

```python
df['mean_ordinal_encode']=df['Cabin'].map(mean_ordinal)
df.head()
```
